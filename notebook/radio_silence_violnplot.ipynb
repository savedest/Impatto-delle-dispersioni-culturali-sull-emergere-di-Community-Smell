## radio silence Globe

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Leggi il file Excel
file_excel = 'C:\\Users\\saver\\PycharmProjects\\Community_smell\\radio_silence\\radio_silence_metrics_globe.xlsx'
df = pd.read_excel(file_excel)

# Nomi delle colonne da analizzare
colonne_da_analizzare = ['CV_1', 'CV_2', 'CV_3', 'CV_4', 'CV_5', 'CV_6', 'CV_7', 'CV_8', 'CV_9']

# Crea una figura con subplots per i Violin Plot
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 17))  # Aumenta l'altezza della figura

# Flattening dell'array di subplots
axs = axs.flatten()

# Crea un Violin Plot per ciascuna colonna
for i, colonna in enumerate(colonne_da_analizzare):
    sns.violinplot(x=df[colonna], inner=None, palette="pastel", ax=axs[i])
    axs[i].set_title(f'Violin Plot di {colonna}')
    axs[i].set_xlabel('Valori')
    axs[i].set_ylabel('Distribuzione')
    axs[i].grid(True)
    
    # Calcola le statistiche: media, mediana e quartili
    mean_val = df[colonna].mean()
    median_val = df[colonna].median()
    quartiles = np.percentile(df[colonna], [25, 75])
    
    # Disegna linee per rappresentare la mediana, la media e i quartili
    axs[i].axvline(x=median_val, color='red', linestyle='--', label=f'Mediana: {median_val:.2f}')
    axs[i].axvline(x=mean_val, color='green', linestyle='--', label=f'Media: {mean_val:.2f}')
    axs[i].axvline(x=quartiles[0], color='blue', linestyle='--', label=f'1° Quartile: {quartiles[0]:.2f}')
    axs[i].axvline(x=quartiles[1], color='purple', linestyle='--', label=f'3° Quartile: {quartiles[1]:.2f}')
    
    # Aggiungi una legenda al subplot corrente
    axs[i].legend(loc='upper right')

# Elimina spazi vuoti per i subplot non utilizzati
for i in range(len(colonne_da_analizzare), len(axs)):
    fig.delaxes(axs[i])

# Ottimizza il layout
plt.tight_layout()

# Mostra il grafico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Leggi il file Excel
file_excel = 'C:\\Users\\saver\\PycharmProjects\\Community_smell\\radio_silence\\radio_silence_metrics_hofstede.xlsx'
df = pd.read_excel(file_excel)

# Nomi delle colonne da analizzare
colonne_da_analizzare = ['CV_pdi', 'CV_idv', 'CV_mas', 'CV_uai', 'CV_ltowvs', 'CV_ivr']

# Crea una figura con subplots per i Violin Plot
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(15, 17))  # Aumenta l'altezza della figura

# Flattening dell'array di subplots
axs = axs.flatten()

# Crea un Violin Plot per ciascuna colonna
for i, colonna in enumerate(colonne_da_analizzare):
    # Rimuovi le righe con valori NaN nella colonna
    colonna_senza_nan = df[colonna].dropna()
    
    sns.violinplot(x=colonna_senza_nan, inner=None, palette="pastel", ax=axs[i])
    axs[i].set_title(f'Violin Plot di {colonna}')
    axs[i].set_xlabel('Valori')
    axs[i].set_ylabel('Distribuzione')
    axs[i].grid(True)
    
    # Calcola le statistiche: media, mediana e quartili
    mean_val = colonna_senza_nan.mean()
    median_val = colonna_senza_nan.median()
    quartiles = np.percentile(colonna_senza_nan, [25, 75])
    
    # Disegna linee per rappresentare la mediana, la media e i quartili
    axs[i].axvline(x=median_val, color='red', linestyle='--', label=f'Mediana: {median_val:.2f}')
    axs[i].axvline(x=mean_val, color='green', linestyle='--', label=f'Media: {mean_val:.2f}')
    axs[i].axvline(x=quartiles[0], color='blue', linestyle='--', label=f'1° Quartile: {quartiles[0]:.2f}')
    axs[i].axvline(x=quartiles[1], color='purple', linestyle='--', label=f'3° Quartile: {quartiles[1]:.2f}')
    
    # Aggiungi una legenda al subplot corrente
    axs[i].legend(loc='upper right')

# Elimina spazi vuoti per i subplot non utilizzati
for i in range(len(colonne_da_analizzare), len(axs)):
    fig.delaxes(axs[i])

# Ottimizza il layout
plt.tight_layout()

# Mostra il grafico
plt.show()
```
